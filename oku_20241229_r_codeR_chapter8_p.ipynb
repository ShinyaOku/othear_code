```r
library(MASS)
library(zeallot)
```

```r
# 相関係数0.7の場合
true_r <- 0.7
rs <- vector("list", 1000)

# rを1000個作成する
n <- 100

# データを生成し、相関係数を計算
for (i in seq_along(rs)) {
  data %<-% MASS::mvrnorm(
    n = n, 
    mu = c(0, 0), 
    Sigma = matrix(c(1, true_r, true_r, 1), ncol = 2))
  c(x, y) %<-% list(data[, 1], data[, 2])
  r <- cor(x, y)
  rs[[i]] <- r
}

# rの分布
hist(unlist(rs), main = "Distribution of Correlation Coefficients", xlab = "r", breaks = 20)

# 統計量Tの算出
ts <- vector("list", 1000)
for (i in seq_along(ts)) {
  t <- (rs[[i]] * sqrt(n - 2)) / sqrt(1 - rs[[i]]^2)
  ts[[i]] <- t
}

# Tの分布
hist(unlist(ts), main = "Distribution of T", xlab = "T", breaks = 20)
```

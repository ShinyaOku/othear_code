```r
library(MASS)
library(zeallot)
```

```r
# 相関係数0.7の場合
true_r <- 0.7
rs <- vector("list", 1000)

# rを1000個作成する
n <- 100

# データを生成し、相関係数を計算
for (i in seq_along(rs)) {
  data %<-% MASS::mvrnorm(
    n = n, 
    mu = c(0, 0), 
    Sigma = matrix(c(1, true_r, true_r, 1), ncol = 2))
  c(x, y) %<-% list(data[, 1], data[, 2])
  r <- cor(x, y)
  rs[[i]] <- r
}

# rの分布
hist(unlist(rs), main = "Distribution of Correlation Coefficients", xlab = "r", breaks = 20)

# Z変換する
zrs <- vector("list", 1000)
for (i in seq_along(zrs)) {
  zr <- 0.5 * log((1 + rs[[i]]) / (1 - rs[[i]]))
  zrs[[i]] <- zr
}

# Z変換したrの分布
hist(unlist(zrs), main = "Distribution of Z-transformed r", xlab = "z-transformed r", breaks = 20)

```

```r
# 95%信頼区間を計算し、true_rがその区間に収まるかチェック
count <- 0
for (i in 1:1000) {
  zr_under95CI <- qnorm(0.025, mean = zrs[[i]], sd = 1 / sqrt(n - 3)) 
  r_under95CI <- (exp(zr_under95CI * 2) - 1) / (exp(zr_under95CI * 2) + 1)
  zr_upper95CI <- qnorm(0.975, mean = zrs[[i]], sd = 1 / sqrt(n - 3)) 
  r_upper95CI <- (exp(zr_upper95CI * 2) - 1) / (exp(zr_upper95CI * 2) + 1)
  
  # true_rが信頼区間内に収まるか確認
  if (r_under95CI <= true_r && true_r <= r_upper95CI) {
    count <- count + 1
  }
  
  # 計算された信頼区間を表示（デバッグ用）
#  print(paste("95% CI for r: ", r_under95CI, " to ", r_upper95CI))
}

# 信頼区間内にtrue_rが収まる回数
print(count)
```
